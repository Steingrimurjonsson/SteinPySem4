```python
import webgetFB
import gzip
import shutil


#1
fb_url = 'https://snap.stanford.edu/data/facebook_combined.txt.gz'
#webgetFB.download(fb_url) 
```

```python
#2
with gzip.open('facebook_combined.txt.gz', 'rb') as file_input:
    with open('facebook_combined.txt', 'wb') as file_output:
        shutil.copyfileobj(file_input, file_output)
```

```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pygraphviz
from networkx.drawing.nx_agraph import graphviz_layout, write_dot
edges = nx.read_edgelist('facebook_combined.txt')
g = nx.Graph(edges)   
print(nx.info(g))

#3 pygraphviz has problems with windows
def draw_graph(graph):
    nx.draw(graph, node_size=30, width=.05, cmap=plt.cm.Blues, 
            with_labels=True, node_color=range(len(graph)))

draw_graph(g)
plt.show()
```

```python
#3:1
graph_dict = dict(g.degree())
sorted_dict = sorted(graph_dict.items(), key=lambda i: i[1], reverse=True)
top_dict = dict(sorted_dict[:10])
print(top_dict)
```
